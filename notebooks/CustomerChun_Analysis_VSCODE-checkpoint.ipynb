## BUILD CHURN DATASET 

```python

def make_churn_dataset(trips, riders, pred_days=60, obs_days=90):
    
    # 1️. Define snapshot date
    max_date = trips["pickup_time"].max()
    snapshot_date = max_date - pd.Timedelta(days=pred_days)

    # 2️. Split into past and future windows
    past_trips = trips[trips["pickup_time"] <= snapshot_date]
    future_trips = trips[
        (trips["pickup_time"] > snapshot_date) &
        (trips["pickup_time"] <= snapshot_date + pd.Timedelta(days=pred_days))
    ]

    # 3️. Create churn label
    future_counts = future_trips.groupby("user_id").size().rename("future_trip_count")
    
    df = riders.merge(future_counts, on="user_id", how="left")
    df["future_trip_count"] = df["future_trip_count"].fillna(0)
    df["churn"] = (df["future_trip_count"] == 0).astype(int)

    # 4️. Recency feature (last trip before snapshot)
    last_trip = past_trips.groupby("user_id")["pickup_time"].max().rename("last_trip_date")
    df = df.merge(last_trip, on="user_id", how="left")

    df["never_used"] = df["last_trip_date"].isna().astype(int)

    df["recency_days"] = (snapshot_date - df["last_trip_date"]).dt.days
    df["recency_days"] = df["recency_days"].fillna(999)

    # 5️. Frequency features
    obs_start = snapshot_date - pd.Timedelta(days=obs_days)
    trips_obs = past_trips[past_trips["pickup_time"] >= obs_start]

    trips_last_30 = past_trips[
        past_trips["pickup_time"] >= snapshot_date - pd.Timedelta(days=30)
    ].groupby("user_id").size().rename("trips_last_30d")

    trips_last_obs = trips_obs.groupby("user_id").size().rename(f"trips_last_{obs_days}d")

    df = df.merge(trips_last_30, on="user_id", how="left")
    df = df.merge(trips_last_obs, on="user_id", how="left")

    df[["trips_last_30d", f"trips_last_{obs_days}d"]] = \
        df[["trips_last_30d", f"trips_last_{obs_days}d"]].fillna(0)

    # 6️. Trend feature
    df["decline_ratio"] = df["trips_last_30d"] / (df[f"trips_last_{obs_days}d"] + 1)

    return df, snapshot_date
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("model", LogisticRegression(class_weight="balanced", max_iter=2000))
])

pipe.fit(X_train, y_train)
```

```python
# Evaluate 
from sklearn.metrics import roc_auc_score, average_precision_score

proba_test = pipe.predict_proba(X_test)[:,1]

roc = roc_auc_score(y_test, proba_test)
pr  = average_precision_score(y_test, proba_test)

print("ROC-AUC:", roc)
print("PR-AUC:", pr)
```

```python
# Tune Threshold
import numpy as np
from sklearn.metrics import f1_score

thresholds = np.linspace(0.01, 0.5, 50)
scores = []

for t in thresholds:
    preds = (proba_test >= t).astype(int)
    scores.append((t, f1_score(y_test, preds)))

best_t, best_f1 = max(scores, key=lambda x: x[1])

print("Best threshold:", best_t)
print("Best F1:", best_f1)
```

```python
from sklearn.metrics import classification_report

final_preds = (proba_test >= best_t).astype(int)
print(classification_report(y_test, final_preds))
```


***Churn current Modelling Insight – Data Analysis***
Our analysis of the RideWise dataset shows that forward-looking churn, defined as “no trips in the next 30–120 days,” is extremely rare, with only ~3.8% of users classified as churned over a 120-day window. Recency and other behavioural features show virtually no correlation with churn (recency mean: churned ≈ 18 days, retained ≈ 17 days), indicating that users who “churn” are nearly as active as those who do not.
This explains why predictive models trained on this definition achieve an AUC close to 0.50 — essentially random performance. The underlying cause is the dataset itself: most users maintain consistent trip activity across months, and there are no structural inactivity patterns to learn from.
**Implication:** Forward-looking churn prediction is not feasible with the current dataset. For meaningful modelling, I recommend defining churn behaviourally (e.g., users inactive for 90 days) to create sufficient signal, increase separation, and allow models to learn predictive patterns.

```python
import pandas as pd

max_date = trips["pickup_time"].max()
snapshot_date = max_date - pd.Timedelta(days=60)

print("Max date:", max_date)
print("New snapshot:", snapshot_date)
```

```python
max_date = trips["pickup_time"].max()
snapshot_date = max_date - pd.Timedelta(days=60)

print("Max date:", max_date)
print("New snapshot:", snapshot_date)
```

```python
# Split Trips Into Past vs Future
past_trips = trips[trips["pickup_time"] <= snapshot_date]
future_trips = trips[
    (trips["pickup_time"] > snapshot_date) &
    (trips["pickup_time"] <= snapshot_date + pd.Timedelta(days=60))
]
```

```python
# Build 60-Day Future Churn Label
future_counts = future_trips.groupby("user_id").size().rename("future_trip_count_60d")

df_label = riders.merge(future_counts, on="user_id", how="left")
df_label["future_trip_count_60d"] = df_label["future_trip_count_60d"].fillna(0)

df_label["churn_next_60d"] = (df_label["future_trip_count_60d"] == 0).astype(int)

print("60-day churn rate:", df_label["churn_next_60d"].mean())
```

```python
# Rebuild Behavioral Features (Using Past Trips Only)
last_trip = past_trips.groupby("user_id")["pickup_time"].max().rename("last_trip_date")

df_model = df_label.merge(last_trip, on="user_id", how="left")

df_model["recency_days"] = (
    snapshot_date - df_model["last_trip_date"]
).dt.days
```

```python
# Trips Last 30 Days (relative to snapshot)
window_30 = snapshot_date - pd.Timedelta(days=30)

trips_30 = past_trips[past_trips["pickup_time"] >= window_30]

freq_30 = trips_30.groupby("user_id").size().rename("trips_last_30d")

df_model = df_model.merge(freq_30, on="user_id", how="left")
df_model["trips_last_30d"] = df_model["trips_last_30d"].fillna(0)
```

```python
# Decline Ratio
window_60 = snapshot_date - pd.Timedelta(days=60)

trips_60 = past_trips[past_trips["pickup_time"] >= window_60]

freq_60 = trips_60.groupby("user_id").size().rename("trips_last_60d")

df_model = df_model.merge(freq_60, on="user_id", how="left")
df_model["trips_last_60d"] = df_model["trips_last_60d"].fillna(0)

df_model["decline_ratio"] = df_model["trips_last_30d"] / (df_model["trips_last_60d"] + 1)
```
